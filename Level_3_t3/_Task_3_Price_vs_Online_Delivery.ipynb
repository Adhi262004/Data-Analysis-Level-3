```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df['has_online_delivery'] = df['has_online_delivery'].apply(lambda x: 1 if str(x).lower() == 'yes' else 0)
df['has_table_booking'] = df['has_table_booking'].apply(lambda x: 1 if str(x).lower() == 'yes' else 0)
```

```python
grouped = df.groupby('price_range')[['has_online_delivery', 'has_table_booking']].mean() * 100
grouped = grouped.round(2).reset_index()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(grouped['price_range'] - 0.15, grouped['has_online_delivery'], width=0.3, label='Online Delivery', color='skyblue')
plt.bar(grouped['price_range'] + 0.15, grouped['has_table_booking'], width=0.3, label='Table Booking', color='salmon')
plt.title("Service Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Percentage Offering Service")
plt.xticks(grouped['price_range'])
plt.legend()
plt.tight_layout()
plt.savefig("Level_3_Task3_ServiceAvailabilityPlot.png")
plt.show()
```
